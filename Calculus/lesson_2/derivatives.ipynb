```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style
sns.set_style("whitegrid")

def derive_plot():
    # Create a figure with three subplots
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
    
    x1 = np.linspace(-2, 2, 1000)
    y1 = x1**2
    dy1 = 2*x1
    
    ax1.plot(x1, y1, label='f(x) = x^2')
    ax1.plot(x1, dy1, '--', label='f\'(x) = 2x')
    ax1.set_title('')
    ax1.grid(True)
    ax1.legend()
    ax1.axhline(y=0, color='k', linestyle='-', alpha=0.3)
    ax1.axvline(x=0, color='k', linestyle='-', alpha=0.3)
    
    x2 = np.linspace(0.1, 4, 1000)
    y2 = np.exp(x2)
    dy2 = np.exp(x2)
    
    ax2.plot(x2, y2, label='f(x) = e^x')
    ax2.plot(x2, dy2, '--', label='f\'(x) = e^x')
    ax2.set_title('')
    ax2.grid(True)
    ax2.legend()
    ax2.axhline(y=0, color='k', linestyle='-', alpha=0.3)
    ax2.axvline(x=0, color='k', linestyle='-', alpha=0.3)
    
    x3 = np.linspace(0.1, 10, 1000)
    y3 = np.sin(x3)
    dy3 = np.cos(x3)
    
    ax3.plot(x3, y3, label='f(x) = sin(x)')
    ax3.plot(x3, dy3, '--', label='f\'(x) = cos(x)')
    ax3.set_title('')
    ax3.grid(True)
    ax3.legend()
    ax3.axhline(y=0, color='k', linestyle='-', alpha=0.3)
    ax3.axvline(x=0, color='k', linestyle='-', alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
def slope_plot():
    # Create a figure with three subplots
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
    
    x1 = np.linspace(-2, 2, 1000)
    y1 = x1**2
    dy1 = 2*x1
    
    ax1.plot(x1, y1, label='f(x) = x^2')
    ax1.plot(x1, dy1, '--', label='f\'(x) = 2x')
    ax1.set_title('')
    ax1.grid(True)
    ax1.legend()
    ax1.axhline(y=0, color='k', linestyle='-', alpha=0.3)
    ax1.axvline(x=0, color='k', linestyle='-', alpha=0.3)
    
    x2 = np.linspace(0.1, 4, 1000)
    y2 = np.exp(x2)
    dy2 = np.exp(x2)
    
    ax2.plot(x2, y2, label='f(x) = e^x')
    ax2.plot(x2, dy2, '--', label='f\'(x) = e^x')
    ax2.set_title('')
    ax2.grid(True)
    ax2.legend()
    ax2.axhline(y=0, color='k', linestyle='-', alpha=0.3)
    ax2.axvline(x=0, color='k', linestyle='-', alpha=0.3)
    
    x3 = np.linspace(0.1, 10, 1000)
    y3 = np.sin(x3)
    dy3 = np.cos(x3)
    
    ax3.plot(x3, y3, label='f(x) = sin(x)')
    ax3.plot(x3, dy3, '--', label='f\'(x) = cos(x)')
    ax3.set_title('')
    ax3.grid(True)
    ax3.legend()
    ax3.axhline(y=0, color='k', linestyle='-', alpha=0.3)
    ax3.axvline(x=0, color='k', linestyle='-', alpha=0.3)
    
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    derive_plot()
```
